A function is a block of reusable code that is defined using def and can be called independently. It is not associated with any specific object and can be used anywhere in the code.

Example of a Function:

def greet(name):

  return f"Hello, {name}!"

print(greet("Alice"))      # Output: Hello, Alice!

greet(name) is a function that takes an argument name and returns a greeting.

It is not tied to an object or class.

A method is similar to a function but is defined inside a class and operates on instances (objects) of that class. Methods always take self as their first parameter (except for class methods and static methods).

Example of a Method:

class Person:

  def __init__(self, name):
        self.name = name

  def greet(self):
        return f"Hello, my name is {self.name}!"

Creating an object of the class

person1 = Person("Bob")

print(person1.greet())              # Output: Hello, my name is Bob!
greet(self) is a method because it is defined inside the class Person.
It operates on an instance (self) and can access instance attributes like self.name.
Key Differences Between Functions and Methods

Definition :	Function is	Defined independently using def whereas Method is	Defined inside a class using def.

Belongs To : Function Belongs To Standalone, not tied to objects whereas method Belongs to a class and operates on instances.

First Parameter : function	Can take any parameters	whereas in case of method the first parameter is usually self (the instance).

Usage	: function is Called directly: function_name() whereas method is	Called on an object: object.method_name()

An iterable is any Python object that can return an iterator.

It contains data that can be iterated over (like lists, tuples, sets, dictionaries, etc.).
It must implement the __iter__() method.
Examples: list, tuple, dict, set, str.
Example of an Iterable

numbers = [1, 2, 3, 4, 5]  # A list (iterable)


for num in numbers:

    print(num)  # Iterating over an iterable

The for loop automatically calls iter(numbers), which returns an iterator.
2. What is an Iterator?
An iterator is an object that produces one value at a time when iterated over.

It must implement two methods:
__iter__() → Returns the iterator object itself.
__next__() → Returns the next value in the sequence and raises StopIteration when finished.
Example of an Iterator

numbers = iter([1, 2, 3, 4, 5])  # Using iter() to get an iterator


print(next(numbers))  # Output: 1

print(next(numbers))  # Output: 2

print(next(numbers))  # Output: 3

Each next() call retrieves the next value.
If no more values exist, StopIteration is raised.
3. Difference Between Iterables and Iterators
Feature	Iterable	Iterator
Definition:	An object that can return an iterator is iterable whereas an object used to iterate over an iterable is iterator.
Methods Required	:__iter__() in iterable	__iter__() and __next__() in iterator.
Examples	list, tuple, set, str, dict of iterables,	iter(list), generators, file objects of iterators.
Memory Usage:	iterables Stores all elements in memory whereas iterators	Generates elements on demand (lazy evaluation)
How to Use	:iterables Can be looped directly whereas iterators	Must use next() manually or in a loop.

You can use the iter() function to get an iterator from an iterable.

5. Creating a Custom Iterator
You can define your own iterator class by implementing __iter__() and __next__().

Example:

class Counter:

    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self  # The iterator object

    def __next__(self):
        if self.current > self.end:
            raise StopIteration  # Stop when reaching end
        self.current += 1
        return self.current - 1  # Return the current value

# Using the iterator
counter = Counter(1, 5)

for num in counter:

    print(num)  # Output: 1 2 3 4 5
The class implements both __iter__() and __next__().
6. Generators: A Simpler Alternative to Iterators

print(next(gen))  # Output: 3

print(next(gen))  # Output: 2

print(next(gen))  # Output: 1

Generators automatically implement __iter__() and __next__().
They do not store all values in memory, making them more efficient than lists.

Generators are a special type of iterator that allow for lazy evaluation, meaning they generate values on the fly instead of storing them in memory. They are memory-efficient and useful for working with large datasets or infinite sequences.

1. Defining a Generator
A generator in Python is defined using a function that contains the yield keyword instead of return.

Example: A Simple Generator

def count_up_to(n):

    count = 1
    while count <= n:
        yield count  # Pauses function and returns value
        count += 1

# Using the generator
gen = count_up_to(5)  # Creates a generator object

print(next(gen))  # Output: 1

print(next(gen))  # Output: 2

print(next(gen))  # Output: 3

Each time next() is called, the function resumes execution from where it left off.
When the function reaches the end, it raises StopIteration.

```python
## QUES 7)What are the advantages of using generators over regular functions?
```

Advantages of Using Generators Over Regular Functions
Generators provide several benefits compared to regular functions, especially when dealing with large datasets or infinite sequences. Below are the key advantages:

1. Memory Efficiency (Lazy Evaluation)
Generators produce values on the fly, meaning they don’t store all values in memory at once.
This is particularly useful for large datasets.

Example: Regular Function vs. Generator
Regular Function (Consumes More Memory)

def squares_list(n):
    return [x ** 2 for x in range(n)]  # Creates a full list in memory

print(squares_list(1000000))  # Stores 1,000,000 values in memory
Generator Function (Memory-Efficient)

def squares_generator(n):
    for x in range(n):
        yield x ** 2  # Generates values one by one

gen = squares_generator(1000000)
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
✅ Advantage: The generator doesn’t store all values at once, reducing memory usage.

2. Improved Performance (Faster Execution)
Generators execute faster because they generate values only when needed, instead of computing and storing everything at once.

Example:

import time

# Regular function
start = time.time()
sum([x ** 2 for x in range(1000000)])  # Creates full list
print("List execution time:", time.time() - start)

# Generator function
start = time.time()
sum(x ** 2 for x in range(1000000))  # Generates values on demand
print("Generator execution time:", time.time() - start)
✅ Advantage: The generator is faster since it doesn’t create an entire list.

3. Can Handle Infinite Sequences
A regular function would require an infinite amount of memory to store an endless sequence.
A generator can handle infinite sequences since it generates values one at a time.

Example: Infinite Number Generator

def infinite_numbers():
    num = 0
    while True:
        yield num
        num += 1

gen = infinite_numbers()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
✅ Advantage: The generator runs indefinitely without consuming excessive memory.

4. Supports Iteration Without Creating Lists
Generators allow iteration over large data sets without creating massive lists in memory.

Example: Processing Large Files

def read_large_file(file_path):
    with open(file_path, "r") as file:
        for line in file:
            yield line  # Reads one line at a time

# Process large file without loading entire file into memory
for line in read_large_file("large_text_file.txt"):
    print(line.strip())  # Prints one line at a time
✅ Advantage: The generator reads one line at a time, preventing memory overflow.

5. Simpler Code (No Need to Manage State)
Generators allow cleaner and more readable code compared to manually managing state with loops.

Example: Fibonacci Sequence
Regular Function (Manually Manages State)

def fibonacci(n):
    fib_seq = []
    a, b = 0, 1
    for _ in range(n):
        fib_seq.append(a)
        a, b = b, a + b
    return fib_seq

print(fibonacci(5))  # Output: [0, 1, 1, 2, 3]
Generator Function (Automatically Resumes Execution)

def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a  # Produces next Fibonacci number
        a, b = b, a + b

gen = fibonacci_generator()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 1
✅ Advantage: The generator remembers its state, making the code cleaner.

6. yield Allows Pause and Resume
Generators pause execution when yield is encountered and resume from the same point when next() is called.

Example: Yielding Data from an API

def fetch_data():
    yield "Data 1"
    yield "Data 2"
    yield "Data 3"

gen = fetch_data()
print(next(gen))  # Output: Data 1
print(next(gen))  # Output: Data 2
✅ Advantage: The function doesn’t have to restart from the beginning each time.

7. Generators Are Iterable (Can Be Used in Loops)
Generators can be used directly in for loops, list comprehensions, and functions like sum().

Example:

def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

print(sum(even_numbers(10)))  # Output: 20 (0+2+4+6+8)
✅ Advantage: Works seamlessly with built-in functions.

The map() function is a built-in Python function that applies a function to every item in an iterable (e.g., list, tuple, set) and returns an iterator with the results.

1. Syntax of map()

map(function, iterable)
function → A function that is applied to each element in the iterable.
iterable → A sequence (list, tuple, etc.) to which the function is applied.
Returns → A map object (which is an iterator).
2. Example: Using map() with a Regular Function

def square(num):
    return num ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
👉 How it works?

The square() function is applied to each element in numbers.
The result is stored in squared_numbers, which is a map object.
list() is used to convert the map object into a list.
3. Example: Using map() with a Lambda Function

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))

print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
✅ Advantage: Using lambda makes the code more concise.

4. Example: Using map() with Multiple Iterables
You can use map() with multiple iterables by passing multiple sequences.


nums1 = [1, 2, 3]
nums2 = [4, 5, 6]

sum_nums = list(map(lambda x, y: x + y, nums1, nums2))

print(sum_nums)  # Output: [5, 7, 9]
👉 Each element from nums1 and nums2 is passed as arguments to the lambda function.

5. Example: Converting Strings to Integers

str_numbers = ["1", "2", "3", "4"]
int_numbers = list(map(int, str_numbers))

print(int_numbers)  # Output: [1, 2, 3, 4]
✅ Advantage: map() helps avoid writing a loop to apply int() to each element.

6. Example: Using map() with Built-in Functions

```python
## QUES 10)What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
```

1. map() Function
✅ Purpose: Applies a function to each element in an iterable and returns an iterator.
Syntax
map(function, iterable)
Example: Squaring Numbers

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
Key Points
Transforms each element in the iterable.
Returns a map object (which is an iterator).
Can take multiple iterables (pairwise operation).
2. filter() Function
✅ Purpose: Filters elements based on a condition, keeping only those for which the function returns True.
Syntax

filter(function, iterable)
Example: Filtering Even Numbers

numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
Key Points
Only selects elements that satisfy a condition.
Returns a filter object (which is an iterator).
Does not modify the elements, just filters them.
3. reduce() Function (from functools module)
✅ Purpose: Repeatedly applies a function to elements and reduces them to a single cumulative result.
Syntax

from functools import reduce
reduce(function, iterable, initial_value)  # `initial_value` is optional
Example: Summing a List

from functools import reduce

numbers = [1, 2, 3, 4]
sum_result = reduce(lambda x, y: x + y, numbers)

print(sum_result)  # Output: 10
Key Points
Combines elements into a single value.
Works cumulatively (e.g., f(f(f(a, b), c), d)).
Needs to be imported from functools.
🔍 Key Differences
Feature	map()	filter()	reduce()
Purpose	Applies a function to each element	Selects elements based on a condition	Combines elements into a single result
Returns	A transformed iterable (map object)	A subset of the original iterable (filter object)	A single reduced value
Function Type	Applies transformation	Applies condition (returns True or False)	Aggregates elements
Modifies Elements?	✅ Yes	❌ No	✅ Yes (combines elements)
Reduces to Single Value?	❌ No	❌ No	✅ Yes
🔹 Example: Using All Three Together
Let's say we have a list of numbers, and we want to:

from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

# Step 1: Filter even numbers
filtered = filter(lambda x: x % 2 != 0, numbers)  # [1, 3, 5]

# Step 2: Square the remaining numbers
squared = map(lambda x: x ** 2, filtered)  # [1, 9, 25]

# Step 3: Sum the squared numbers
result = reduce(lambda x, y: x + y, squared)  # 1 + 9 + 25 = 35

print(result)  # Output: 35

Use map() for transformations.
Use filter() for selecting specific values.
Use reduce() for aggregation into a single result.

Step 1: Understanding reduce()

The reduce() function applies a given function cumulatively to the elements of an iterable, reducing them to a single value.

Syntax:

from functools import reduce

reduce(function, iterable)

The function takes two arguments.
The first two elements are passed to the function first.
The result of the function is used as the first argument for the next element, and so on.

Step 2: Defining the Sum Function

We use a lambda function for summation:

lambda x, y: x + y

This function takes two numbers, adds them, and returns the result.

Step 3: Applying reduce() to [47, 11, 42, 13]

from functools import reduce

numbers = [47, 11, 42, 13]

result = reduce(lambda x, y: x + y, numbers)

print(result)  # Output: 113

Step-by-Step Execution
Iteration	x (Accumulated Result)	y (Next Element)	Computation (x + y)	Output

1st	47	11	47 + 11 = 58	58

2nd	58	42	58 + 42 = 100	100

3rd	100	13	100 + 13 = 113	113

Step 4: Final Result
After the last iteration, the final reduced value is 113.

Thus, the sum of [47, 11, 42, 13] using reduce() is:
113

```python
## QUES 4) Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Checking prime numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
```

```python
## QUES 5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
## terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Number of terms to generate
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Keeps track of iterations

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration  # Stops iteration when limit is reached

        # Generate the next Fibonacci number
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update values
        self.count += 1
        return fib_number

fib_iter = FibonacciIterator(10)  # Create an iterator for 10 Fibonacci numbers

for num in fib_iter:
    print(num, end=" ")
```

```python
## QUES 6) Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):  # Loop from 0 to max_exponent
        yield 2 ** exponent  # Yield the power of 2

# Example Usage
for value in powers_of_two(5):  # Generates 2^0 to 2^5
    print(value, end=" ")
```

```python
## QUES 7) Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:  # Open file in read mode
            for line in file:
                yield line.strip()  # Yield each line after stripping whitespace
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage
file_path = "sample.txt"  # Replace with actual file path

for line in read_file_line_by_line(file_path):
    print(line)  # Prints each line from the file



```

```python
## QUES 9) Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
##Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32  # Formula: (°C × 9/5) + 32 = °F

# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Using map() to convert temperatures
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Output result
print(fahrenheit_temps)
```

```python
## QUES 10)  Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = "aeiouAEIOU"  # Define vowels
    return "".join(filter(lambda char: char not in vowels, string))  # Filter out vowels

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```
